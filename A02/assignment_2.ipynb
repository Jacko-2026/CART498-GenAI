```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# The poem as input
poem = """
One must have a mind of winter
To regard the frost and the boughs
Of the pine-trees crusted with snow;
And have been cold a long time
To behold the junipers shagged with ice,
The spruces rough in the distant glitter
Of the January sun; and not to think
Of any misery in the sound of the wind,
In the sound of a few leaves,
Which is the sound of the land
Full of the same wind
That is blowing in the same bare place
For the listener, who listens in the snow,
And, nothing himself, beholds
Nothing that is not there and the nothing that is.
"""

# Split the poem into lines
lines = poem.strip().split('\n')

# Function to get the seventh-highest word prediction
def get_seventh_word(line):
    # Encode the input line
    inputs = tokenizer.encode(line, return_tensors="pt")

    # Get predictions for the next word
    with torch.no_grad():
        outputs = model(inputs)
        predictions = outputs.logits[:, -1, :]

    # Get the top 7 predicted words
    top_7 = torch.topk(predictions, 7, dim=-1).indices[0].tolist()

    # Map token ids back to words
    top_7_words = [tokenizer.decode([word_id]) for word_id in top_7]

    return top_7_words[6]  # Return the 7th word

# Process the poem to create P+7 version
def process_poem(lines):
    processed_lines = []

    for line in lines:
        words = line.split()
        last_word = words[-1]

        # Get the seventh-highest probability word to replace the last word
        seventh_word = get_seventh_word(line)

        # Replace last word with seventh word
        new_line = ' '.join(words[:-1]) + ' ' + seventh_word
        processed_lines.append(new_line)

    return '\n'.join(processed_lines)

# Create the P+7 version
p7_version = process_poem(lines)


# Save both versions as .txt files
with open("P+7.txt", "w") as f:
    f.write(p7_version)


```
